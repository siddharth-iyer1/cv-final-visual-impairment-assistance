```python
import cv2
import matplotlib.pyplot as plt
```

```python
images  = []

camera1_pic = cv2.VideoCapture(0)
camera2_pic = cv2.VideoCapture(1)

ret1, frame1 = camera1_pic.read()
if ret1:
    frame_rgb1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
    images.append(frame_rgb1)
    cv2.imshow('camera 1', frame_rgb1)
else:
    print('camera 1 not found')

ret2, frame2 = camera2_pic.read()
if ret2:
    frame_rgb2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB)
    images.append(frame_rgb2)
    cv2.imshow('camera 2', frame_rgb2)
else:
    print('camera 2 not found')

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
for i, img in enumerate(images):
    axs[i].imshow(img)
    axs[i].set_title(f'Image {i+1}')
plt.show()
```
